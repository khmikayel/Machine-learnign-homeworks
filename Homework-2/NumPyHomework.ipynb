### 9

Write a function plot_fun(f,xmin, xmax): that takes as an argument a function f(x), and numbers xmin and xmax and plots the graph of the function in the region of xmin to xmax.

If you want to test this you need to pass a FUNCTION instead of f and not an expression such as x**2 + x.<br>
For example<br>
def my_fun(x):<br>
....return x**2 + x + 3*np.sin(x)<br>
<br>
Now you can pass call plot_fun(my_fun, 1, 10) to test the plot_fun() function with your function my_fun()

```python
def plot_fun(fun, x_min, x_max):
   
    x = np.linspace(x_min, x_max, 1000)
    y = np.empty(1000)
    
    for i in range(1000):
        y[i] = fun(x[i])
    
    plt.plot(x, y)

def my_fun(x):
    
    if x == 0:
        return 0
    else:
        return np.e**(-1 / x**2)
    
x_min = -15
x_max = 15

plot_fun(my_fun, x_min, x_max)
```

### 11
Do what you did in 9, but now plot 20 randomly selected points not exactly on the function f(x)'s graph, the y values of the points should have some error from the function. The vertical distance of the pints from the graph of f(x) should be from a normal random distribution with some standard deviation sigma. Vary the sigma value to see what you get.

```python
def my_fun_error(x, mu, sigma):
    return my_fun(x)+np.random.normal(mu, sigma)

x = x_min + (x_max - x_min)*np.random.rand(20)
y = np.empty(20)
z = np.empty(20)
t = np.empty(20)

for i in range(20):
    y[i] = my_fun_error(x[i], 0, 1)
    z[i] = my_fun_error(x[i], 0, 0.01)
    t[i] = my_fun_error(x[i], 0, 1.5)

plot_fun(my_fun, x_min, x_max)
plt.scatter(x, y, marker='o', color='red')
plt.scatter(x, z, marker='o', color='black')
plt.scatter(x, t, marker='o', color='yellow')
```

### 12.* 
Make a smoother image by applying a filter 
- hint: use a convolution

```python
%pylab inline
img = mpimg.imread('noisy.png')
imgplot = plt.imshow(img)
plt.show()
```

```python
img.shape
```

```python
filter = np.array([[1, 4,  7,  4,  1],
                   [4, 16, 26, 16, 4],
                   [7, 26, 41, 26, 7],
                   [4, 16, 26, 16, 4],
                   [1, 4,  7,  4,  1]])

filter = (1/273)*filter

def apply_filter(arr):
    return np.sum(filter * arr)

def filter_image(im):
    
    filtered_im = np.zeros((800, 800, 3))
    
    for level in range(3):
        for i in range(2,798):
            for j in range(2,798):
                cell_neighborhood = im[i-2 : i+3, j-2 : j+3, level]
                filtered_im[i, j, level] = apply_filter(cell_neighborhood)
    
    return filtered_im

for i in range(13):
    img = filter_image(img)
```

```python
plt.imshow(img)
imgplot = plt.imshow(img)
```

```python
matplotlib.image.imsave('noise_free.png', img)
```
